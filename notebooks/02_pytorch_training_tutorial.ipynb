```python
import torch
import numpy as np

data = [[1, 2],
        [3, 4]]
x_data = torch.tensor(data=data)
print(f"From data: {x_data}")

np_array = np.array(data)
x_np = torch.from_numpy(np_array)
print(f"From numpy array: {x_np}")

#tensor of 1s like from another tensor(retaining the properties)
x_ones = torch.ones_like(x_data) #retain the properties of x_data (shape, datatype)
print(x_ones)
#tensor of random from another tensor, overriding properties
x_rand = torch.rand_like(x_data, dtype=torch.float) #overrides the properties of x_data
print(x_rand)
```

```python
shape = (2, 3,) #Tuple of tensor dimensions (2 rows, 3 columns)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(rand_tensor)
print(ones_tensor)
print(zeros_tensor)
```

```python
tensor = torch.rand(3, 4)

print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

```python
#Moving tensor to the GPU if available
device = torch.accelerator.current_accelerator().type if torch.accelerator.is_available() else 'cpu'
tensor = tensor.to(device=device)
print(f"Tensor is stored on: {tensor.device}")
```

```python
tensor = torch.ones(4, 4,)
tensor[:,1] = 0 #butun satirlari, ikinci sutundaki 0 yap
print(tensor)
print()

#element wise multiplication:
tensor_2 = tensor
tensor_2[:,2] = 0
tensor = tensor.mul(tensor_2)
tensor = tensor * tensor_2
print(tensor)
```

```python
t = torch.ones(5)
print(f"t={t}")
n = t.numpy()
print(f"n={n}")

t.add_(5)
print(f"t={t}")
print(f"n={n}")
```

```python
n = np.ones(5)
print(f"n={n}")
t = torch.from_numpy(n)
print(f"t={t}")

np.add(n, 1, out=n)
print(f"n={n}")
print(f"t={t}")
```
